```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Importing dataset
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
# Displaying first 5 rows 
df.head()
```

```python
df.describe()

```

```python
df = df[df["Publishing Year"]> 1900]
```

```python
# Checking N/A Values
df.isna().sum()
```

```python
# Drop N/A Book Names
df.dropna(subset = "Book Name", inplace = True)
```

```python
# Checking for duplicated rows
df.duplicated().sum()
```

```python
#Check unique values
df.nunique()
```

```python
# Creating Histogram of Publishing year

plt.hist(df['Publishing Year'])
plt.xlabel("Publishing Year")
plt.ylabel('Frequency')
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
#Create Bar Chart for Number of books in each genre
df['genre'].value_counts().plot(kind = 'bar')
plt.xlabel("Genre")
plt.ylabel('Number of Books ')
plt.title("Number of books in each genre")
plt.show()
```

```python
# Grouping rating by author
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending= False)
```

```python
# Boxplot
sns.boxplot(x = 'genre' , y = 'Book_ratings_count', data = df)
plt.xlabel("genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot of Book rating count for Each Genre")
plt.show()
```

```python
# ScatterPlot Sale price and units sold

plt.scatter(df['sale price'], df['units sold'])
plt.xlabel('Sale Price')
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units Sold")
plt.show()

# When sale price drops units sold Increases 
```

```python
# Get value counts of language code
language_counts = df['language_code'].value_counts()
```

```python
# Create PieChart
plt.pie(language_counts, labels = language_counts.index, startangle = 90, autopct="%1.1f%%" )
plt.title('Language distribution of books')
plt.show()
```

```python
#Total rev  per publisher
df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending = False)
```

```python
# Group data by author rating find average book ratings count for each author

df.groupby("Author_Rating")['Book_ratings_count'].mean().sort_values(ascending=False).max()
```

```python
df.groupby("language_code").size().sort_values(ascending=False)
```

```python
#Variance of author ratings

df.groupby('Author_Rating')["Book_ratings_count"].var()
```

```python
#Scatterplot Average book ratings

plt.scatter(df['Book_average_rating'], df['Book_ratings_count'])
plt.xlabel('Book Average Rating')
plt.ylabel('Book Ratings Count')
plt.title("Scatter Plot of Book Average Rating vs Book Ratings count")
plt.show()
```

```python
total_gross_sales_by_autor = df.groupby("Author")['gross sales'].sum()
total_gross_sales_by_autor.sort_values(ascending=False).head(15).plot(kind= 'bar')
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
plt.show()
```

```python
#Boxplot of Units sold for all author rating

sns.boxplot(x = 'Author_Rating', y = 'units sold' , data = df)
plt.xlabel("Author Rating")
plt.ylabel('Units Sold')
plt.title('Boxplot of Units sold for each Author Rating')
plt.show()

```

```python
df.groupby('Publishing Year')['units sold'].sum().plot(kind = 'line' , marker = 'o')
plt.xlabel('Publishing Year')
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold of the Years")
plt.show()
```
